```python
import re
def parse_line(line):
    """
    解析一行数据，提取时间戳、ADXL1、ADXL2、AN1、AN2。
    
    参数:
        line (str): 一行原始数据字符串。
        
    返回:
        dict: 包含 'timestamp', 'ADXL1_x', 'ADXL1_y', 'ADXL1_z',
              'ADXL2_x', 'ADXL2_y', 'ADXL2_z', 'AN1', 'AN2' 的字典。
              如果解析失败，返回 None。
    """
    # 正则表达式匹配整行
    pattern = (
        r'(?P<timestamp>\d{2}:\d{2}:\d{2}\.\d{3})>.*?'
        r'ADXL1:(?P<ADXL1_x>-?\d+\.?\d*),(?P<ADXL1_y>-?\d+\.?\d*),(?P<ADXL1_z>-?\d+\.?\d*).*?'
        r'ADXL2:(?P<ADXL2_x>-?\d+\.?\d*),(?P<ADXL2_y>-?\d+\.?\d*),(?P<ADXL2_z>-?\d+\.?\d*).*?'
        r'AN:(?P<AN1>\d+).*?AN:(?P<AN2>\d+)'
    )
    
    match = re.search(pattern, line)
    if not match:
        return None
    
    data = match.groupdict()
    # 转换为适当的数据类型
    result = {
        'timestamp': data['timestamp'],
        'ADXL1_x': float(data['ADXL1_x']),
        'ADXL1_y': float(data['ADXL1_y']),
        'ADXL1_z': float(data['ADXL1_z']),
        'ADXL2_x': float(data['ADXL2_x']),
        'ADXL2_y': float(data['ADXL2_y']),
        'ADXL2_z': float(data['ADXL2_z']),
        'AN1': int(data['AN1']),
        'AN2': int(data['AN2'])
    }
    return result
```

```python
import csv

def process_log_to_csv(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as fin, \
         open(output_file, 'w', newline='', encoding='utf-8') as fout:
        
        fieldnames = ['timestamp', 'ADXL1_x', 'ADXL1_y', 'ADXL1_z',
                      'ADXL2_x', 'ADXL2_y', 'ADXL2_z', 'AN1', 'AN2']
        writer = csv.DictWriter(fout, fieldnames=fieldnames)
        writer.writeheader()
        
        for line in fin:
            parsed = parse_line(line.strip())
            if parsed:
                writer.writerow(parsed)
```

```python
for k,v in data_config.items():
    process_log_to_csv(v, f'../data/processed/{k}.csv')
```
